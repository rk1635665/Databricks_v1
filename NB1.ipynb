```python
from databricks.sdk import AccountClient

# Initialize client (requires Account ID and authentication)
a = AccountClient()

# List all metastores in the account
metastores = list(a.metastores.list())

if not metastores:
    print("No metastores found in this account.")
else:
    for m in metastores:
        print(f"Name: {m.name}, ID: {m.metastore_id}, Region: {m.region}")

# Specifically check if the current workspace is assigned to one
workspace_id = dbutils.notebook.entry_point.getDbutils().notebook().getContext().workspaceId().get()
try:
    assignment = a.metastore_assignments.get(workspace_id=int(workspace_id))
    print(f"Current Workspace is assigned to Metastore ID: {assignment.metastore_id}")
except Exception as e:
    print("Workspace is not assigned to a Unity Catalog metastore.")
```
